```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from skimage.exposure import match_histograms

```

```python
# Load images
source = cv2.imread("C:/Users/sudob/OneDrive/Desktop/lab5.ipynb/images (2).jpg")
target = cv2.imread("C:/Users/sudob/OneDrive/Desktop/lab5.ipynb/images.jpg")

source_rgb = cv2.cvtColor(source, cv2.COLOR_BGR2RGB)
target_rgb = cv2.cvtColor(target, cv2.COLOR_BGR2RGB)

plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Source Image")
plt.imshow(source_rgb)
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Target Image")
plt.imshow(target_rgb)
plt.axis('off')
```

```python
# Convert source to grayscale
source_gray = cv2.cvtColor(source, cv2.COLOR_BGR2GRAY)

# Apply histogram equalization
equalized_gray = cv2.equalizeHist(source_gray)

# Show results
plt.figure(figsize=(10,5))

plt.subplot(1,2,1)
plt.title("Original Grayscale")
plt.imshow(source_gray, cmap='gray')
plt.axis('off')

plt.subplot(1,2,2)
plt.title("Equalized Grayscale")
plt.imshow(equalized_gray, cmap='gray')
plt.axis('off')
```

```python
# Equalize each channel
equalized_color = np.zeros_like(source)

for i in range(3):
    equalized_color[:,:,i] = cv2.equalizeHist(source[:,:,i])

equalized_color_rgb = cv2.cvtColor(equalized_color, cv2.COLOR_BGR2RGB)

# Show result
plt.figure(figsize=(5,5))
plt.title("Equalized Color Image")
plt.imshow(equalized_color_rgb)
plt.axis('off')
```

```python
# Match histogram of equalized image with target image
matched = match_histograms(equalized_color_rgb, target_rgb, channel_axis=-1)

# Convert to uint8
matched = np.clip(matched, 0, 255).astype(np.uint8)

# Show result
plt.figure(figsize=(5,5))
plt.title("Histogram Matched Image")
plt.imshow(matched)
plt.axis('off')
```

```python
plt.figure(figsize=(12,8))

plt.subplot(2,2,1)
plt.title("Original Source")
plt.imshow(source_rgb)
plt.axis('off')

plt.subplot(2,2,2)
plt.title("Equalized Image")
plt.imshow(equalized_color_rgb)
plt.axis('off')

plt.subplot(2,2,3)
plt.title("Target Image")
plt.imshow(target_rgb)
plt.axis('off')

plt.subplot(2,2,4)
plt.title("Matched Image")
plt.imshow(matched)
plt.axis('off')
```

```python
plt.figure(figsize=(10,6))

plt.hist(source_rgb.ravel(), bins=256, alpha=0.5, label="Source")
plt.hist(equalized_color_rgb.ravel(), bins=256, alpha=0.5, label="Equalized")
plt.hist(matched.ravel(), bins=256, alpha=0.5, label="Matched")

plt.legend()
plt.title("Histogram Comparison")
plt.xlabel("Pixel Intensity")
plt.ylabel("Frequency")
plt.show()
```
